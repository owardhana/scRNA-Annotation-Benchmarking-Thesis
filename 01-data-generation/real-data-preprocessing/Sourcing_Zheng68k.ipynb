```python
import scvelo as scv
import scipy.io as sio
import pandas as pd
import os

# 1. Load the AnnData object
adata = scv.datasets.pbmc68k(file_path='data/PBMC/pbmc68k.h5ad')

# Create a folder to keep things organized
out_dir = "./"
os.makedirs(out_dir, exist_ok=True)

# ==============================================================================
# 2. Extract and Write the Counts Matrix
# ==============================================================================
# AnnData is [Cells, Genes]. Seurat expects [Genes, Cells].
# We use .T to transpose the sparse matrix before writing!
counts_matrix = adata.X.T 
sio.mmwrite(f"{out_dir}counts.mtx", counts_matrix)

# ==============================================================================
# 3. Extract and Write Barcodes (Cells)
# ==============================================================================
# obs_names contains the cell barcodes
barcodes = pd.DataFrame(adata.obs_names)
barcodes.to_csv(f"{out_dir}barcodes.tsv", sep='\t', index=False, header=False)

# ==============================================================================
# 4. Extract and Write Genes (Features)
# ==============================================================================
# var_names contains the gene symbols/IDs
genes = pd.DataFrame(adata.var_names)
genes.to_csv(f"{out_dir}genes.tsv", sep='\t', index=False, header=False)

# ==============================================================================
# 5. Extract and Write Metadata
# ==============================================================================
# obs contains all the cell-level metadata (clusters, tSNE coords, etc.)
# We keep the index (barcodes) so Seurat can map it correctly
adata.obs.to_csv(f"{out_dir}metadata.tsv", sep='\t', index=True, index_label="barcode")

print(f"Extraction complete! Files saved to {out_dir}")
```
